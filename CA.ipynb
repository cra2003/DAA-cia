```python
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split


df = pd.read_csv(r"Users/cra/Downloads/Bank_Personal_Loan_Modelling.csv")

x = df.loc[:, df.columns != 'Personal Loan'].values
y = df.loc[:,df.columns == 'Personal Loan'].values
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(13,)),
    tf.keras.layers.Dense(32, activation='relu',),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```python

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy()
optimizer = tf.keras.optimizers.Adam()

model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
model.fit(x, y, epochs=5, batch_size=32, validation_data=(x, y))
```

```python
n_pop = 50 
n_gen = 200 
alpha = 0.25 
beta = 0.25 
p_c = 0.8 
p_m = 0.1 
```

```python
population = []
for i in range(n_pop):
    population.append(model.get_weights())
```

```python
for gen in range(n_gen):
    # Evaluate fitness
    fitness = []
    for i in range(n_pop):
        model.set_weights(population[i])
        y_pred = model.predict(x)
        loss = loss_fn(y, y_pred)
        fitness.append(1 / (1 + loss.numpy()))
    fitness = np.array(fitness)

    # Selection
    fitness_sum = np.sum(fitness)
    p_select = fitness / fitness_sum
    selected_idx = np.random.choice(n_pop, size=n_pop, p=p_select)

    # Crossover
    offspring = []
    for i in range(n_pop):
        if np.random.rand() < p_c:
            parent1 = selected_idx[i]
            parent2 = np.random.choice(selected_idx)
            child = []
            for w1, w2 in zip(population[parent1], population[parent2]):
                if np.random.rand() < beta:
                    w = alpha * w1 + (1 - alpha) * w2
                else:
                    w = w1
                child.append(w)
            offspring.append(child)
        else:
            offspring.append(population[selected_idx[i]])

    # Update population
    population = offspring
```

```python
# Set the final weights to the best individual
best_idx = np.argmax(fitness)
best_weights = population[best_idx]
model.set_weights(best_weights)
```
